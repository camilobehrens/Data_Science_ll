# 8 Análisis Uni y Bivariado

## 8.1 Análisis Univariado

### Introducción

En esta unidad se explorarán los conceptos esenciales del análisis estadístico, enfocados enel análisis univariado y bivariado, que son pilares fundamentales en la exploración e interpretación de datos. Estos métodos son utilizados para describir, visualizar y análizar patrones dentro de conjuntos de datos, proporcionando una base sólida para la toma de decisiones basadas en evidencia.

### Análisis Univariado

**Definición** 

Se refiere al estudio de una sola variables. Se centra en la descripción y resumen de los datos mediante estadísticas descriptivas. Este tipo de análisis es fundamental para comprender las características básicas de una variable antes de proceder con análisis más complejos.

**Medidas de Tendencia Central**

Las medidas de tendencia central son estadísticas que indican el valor central o más típico en un conjunto de datos. Las principales son:

* *Media:* el promedio de todos los valores de la variable.
* *Mediana:* el valor que divide el conjunto de datos en dos mitades iguales. Es más robusta ante valores extremos o atípicos.
* *Moda:* el valor que aparece con más frecuancia en el conjunto de datos.


**Medidas de Disperción**

Las medidas de dispersión proporcionan información sobre la variabilidad o el grado de dispersión de los datos con respecto a la medida de tendencia central.

* *Varianza:* indica que tan dispersos están los valores en relación con la media. Es el promedio de las diferencias al cudrado entre los datos y la media.

* *Desviación estándar:* es la raíz cuadrada de la varianza y se utiliza para medir la dispersión de los datos en las mismas unidades que la variable original.


**Gráficos Comunes**

Para visualizar el comportamiento de una variable en un análisis univariado los gráficos más comunes son:


* *Histogramas:* muestran la distribución de los datos dividiéndolos en intervalos y representando la frecuencia de los valores en cada intervalo.

* *Box Plots:* también conocidos como diagramas de caja, muestran la mediana, los cuartiles y los valores atípicos, ofreciendo una visión clara de la dispersión y la simetría de los datos.


Este conjunto de herramientas y medidas permite explorar en profundidad las características de una única variable, proporcionando una base sólida para futuros análisis.

## 8.2 Análisis Bivariado

### Definición 

Se refiere al estudio de la relación entre dos conjuntos de datos, con el objetivo de identificar si existe alguna asociación o dependencia entre ellos. Este análisis se utiliza para comprender cómo una variable influye o está relacionada con otra.


### Tipos de Análisis

En el análisis bivariado, las relaciones pueden clasificarse en función del tipo de variable involucrada:

* **Numérico vs Numérico:** ambas variables son numéricas y se pueden analizar utilizando herramientas como la correlación o la regresión lineal para determinar la relación entre ellas.

* **Categórico vs Cetegórico:** ambas variables son categóricas. En este caso se pueden analizar utilizando tablas de contingencia o medidas como el coeficiente Phi para evaluar la asociación entre las variables.

* **Numérico vs Categórico:** una variable es numérica y la otra es categórica. Este tipo de análisis se puede abordar mediante comparaciones de medias o pruebas estadísticas que contrasten los grupos categóricos en relación con la variable numérica.


### Gráficos Comunes

Para visualizar la relación entre dos variables en un análisis bivariado los gráficos más comunes son:

* **Diagrama de dispersión (Scatterplot):** muestra la relación entre dos variables numéricas y ayuda a identificar patrones, tendencias o correlaciones entre ellas.


Este conjunto de herramientas y métodos permiten explorar y visualizar la interacción entre dos variables, aportando información valiosa para la toma de decisiones.
